# Introduction

The Krackhardt kite, named after researcher David Krackhardt, is characterized by containing 10 nodes and 18 edges. In graph theory, a node represents an individual, while edges represents the relationships and direction amongst each individual.

To recreate the Krackhardt kite I will be utilizing the library *networkx* and *matplotlib*. 

# Solution 1

In the first solution, I will initialize our graph using *nx.krackhardt_kite_graph()*. This will automatically load a chart preloaded with the nodes and edges.

```python
# Import libraries
import networkx as nx
import matplotlib.pyplot as plt

# Create krackhardt kite graph
G = nx.krackhardt_kite_graph()
```

I then create a dictionary object with the names from the example to each node.

```python
new_labels = {
    0: 'Beverly',
    1: 'Andre',
    2: 'Ed',
    3: "Diane",
    4: "Carol",
    5: "Garth",
    6: "Fernando",
    7: "Heather",
    8: "Ike",
    9: "Jane",
}
```

Using *plt.subplots()*, I will create a figure and axes object which we will pass to the *nx.draw()* function so we can visualize the graph.

```python
# Create a matplotlib figure and axes
fig, ax = plt.subplots()

# Visualize the graph
nx.draw(G, with_labels=True, labels=new_labels, node_color='orange', node_size=1000, font_size=6, font_weight='bold', font_color='blue', ax=ax)

# Show the plot
plt.show()
```

# Solution 2

In the second solution, I will create the graph object from scratch by starting with *nx.graph()* instead of *nx.krackhardt_kite_graph()*.

In addition, I will manually create the edges to the object. To find out what numbers to use to create the edges, I will examine object G (which contains the Krackhardt kite graph) using the *G.edges()* function.

```python
# Create empty graph object
g = nx.Graph()

# Check edges from krackhardt object
G.edges()

# Manually create edges
e = [(0, 1), (0, 2), (0, 3), (0, 5), (1, 3), (1, 4), (1, 6), (2, 3), (2, 5), (3, 4), (3, 5), (3, 6), (4, 6), (5, 6), (5, 7), (6, 7), (7, 8), (8, 9)]

# Add edges to empty graph object
g.add_edges_from(e)
```

Here I manually assign the node labels from earlier, and graph object *g* into a new object: *k*

```python
k = nx.relabel_nodes(g,new_labels)
```

```python
fig, ax = plt.subplots()
nx.draw(k, with_labels=True, ax=ax)
plt.show()
```
